Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\JINDU\Documents\Customer Call List.xlsx')
df
```

```python
# Removing duplicates and saving

df = df.drop_duplicates()
df
```

```python
# Removing unused column

df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
# USing strip to remove elemnt in last name abd saving to only lastName column

#df['Last_Name'] = df['Last_Name'].str.lstrip('...')
#df['Last_Name'] = df['Last_Name'].str.lstrip('/')
#df['Last_Name'] = df['Last_Name'].str.rstrip('_')

#Put every element into the pparenthesis
df['Last_Name'] = df['Last_Name'].str.strip('./_')

df
```

```python
# Cleaning Phone number
df['Phone_Number'] = df['Phone_Number'].str.replace('|', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('-', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('/', '')
df['Phone_Number']
```

```python
# Formatting phone number xxx-xxx-xxxx
# First converting Phone_number to str because of NaN
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
# Formatting...
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# Removing NaN
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

```python
# Spliting Address column into Street address, State and ZipCode
df[['Street_Address','State','Zip_Code']] = df['Address'].str.split(',',expand=True)
df
```

```python
# Standardizing Paying customer column
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
# Standardizing Paying Do_Not_Contact column
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
# Replacing all NaN Value
#df.replace('N/a','')
#df.replace('NaN','')
df = df.fillna('')
df
```

```python
# Let remove the rows of people that do not want calls
# df.dropna(subset='Phone_Number', inplace=True)

for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
# Let remove the rows of people that do not have Phone Number
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
df
```

```python
# Reseting Index
df = df.reset_index(drop=True)
df
```
